```python
numpad = [" " for x in range(0,10)]

def board(numpad):
    print('{} | {} | {}'.format(numpad[7], numpad[8], numpad[9]))
    print('{} | {} | {}'.format(numpad[4], numpad[5], numpad[6]))
    print('{} | {} | {}'.format(numpad[1], numpad[2], numpad[3]))

def player_marker(player1):
    if player1 == 'X':
        return 'O'
    else:
        return 'X'

def player_choice(numpad, player):
    choice = None
    while choice not in range(1, 10) or numpad[choice] != " ":
        choice = int(input(f"Player {player}, choose a number between 1 and 9 to place your marker: "))
    return choice

def victory_check(numpad, player):
    if (numpad[7] == numpad[8] == numpad[9] == player or
        numpad[4] == numpad[5] == numpad[6] == player or
        numpad[1] == numpad[2] == numpad[3] == player or
        numpad[7] == numpad[5] == numpad[3] == player or
        numpad[9] == numpad[5] == numpad[1] == player or
        numpad[7] == numpad[4] == numpad[1] == player or
        numpad[8] == numpad[5] == numpad[2] == player or
        numpad[9] == numpad[6] == numpad[3] == player):
        return True
    else:
        return False

def draw_check(numpad):
    for i in range(1, 10):
        if numpad[i] == " ":
            return False
    return True

# Choose player marker
player1 = ""
while player1 != 'X' and player1 != 'O':
    player1 = input("Please choose 'X' or 'O' for marker: ")
player2 = player_marker(player1)

# Game loop
game_over = False
players = [player1, player2]
turn = 0
while not game_over:
    print(f"It's {players[turn]}'s turn")
    choice = player_choice(numpad, turn + 1)
    numpad[choice] = players[turn]
    board(numpad)
    if victory_check(numpad, players[turn]):
        print(f"Player {turn + 1} wins!")
        game_over = True
    elif draw_check(numpad):
        print("It's a draw.")
        game_over = True
    else:
        turn = (turn + 1) % 2
```
